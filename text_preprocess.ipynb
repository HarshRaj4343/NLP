# Removing HTML tags

```python
import re

def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub(r'', text)

# Applying it to a DataFrame column
df['review'] = df['review'].apply(remove_html_tags)
```

# Removing URLs

```python
import re

def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)

# Applying it to a text
sample_text = "Check out this link: https://example.com"
clean_text = remove_url(sample_text)
```

```python
df['review'] = df['review'].apply(remove_url)
```

# Removing Punctuations - M1 


```python
import string,time
exclusion = string.punctuation
```

```python
def remove_punc(text):
    for char in exclusion:
        text = text.replace(char,'')
    return text

text = 'string. With. Punctuation?'
clean_text = remove_punc(text)
print(clean_text)
```

```python
df['review'] = df['review'].apply(remove_punc)
```

```python
df
```

# Removing Punctuations - M2


```python
def remove_punc1(text):
    return text.translate(str.maketrans('', '', exclusion))

text = 'string. With. Punctuation?'
clean_text = remove_punc(text)
print(clean_text)
```

# Fixing Chat Words

```python
chat_words = {
    "AFAIK": "As Far As I Know",
    "ASAP": "As Soon As Possible",
    "BRB": "Be Right Back",
    "GN": "Good Night"
}

def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)

text = "AFAIK ,I'll BRB after this meeting. GN ! "
clean_text = chat_conversion(text)
print(clean_text)
```

```python
df['review'] = df['review'].apply(chat_conversion)
```

# Stop Words Removal

```python
from nltk.corpus import stopwords
import nltk

nltk.download('stopwords')
stop_words = stopwords.words('english')

def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word not in stop_words:
            new_text.append(word)
    return " ".join(new_text)
```

```python
df['review'] = df['review'].apply(remove_stopwords)
```

```python
import re
def remove_emoji(text):
    emoji_pattern = re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    return emoji_pattern.sub(r'', text)
```

# Tokenisation

## Using NLTK

```python
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk

# Downloading the required NLTK data files (punkt)
nltk.download('punkt')

dummy_text = "This is a dummy text. It contains multiple sentences for tokenization."

# 1. Sentence Tokenization (Splits the paragraph into a list of sentences)
sentences = sent_tokenize(dummy_text)
print("Sentences:", sentences)

# 2. Word Tokenization (Splits the sentences into individual words)
for sentence in sentences:
    words = word_tokenize(sentence)
    print("Words:", words)
```

# Stemming


```python
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
def stem_words(text):
    return " ".join([ps.stem(word) for word in text.split()])
```

```python
df['review'] = df['review'].apply(stem_words)
```
