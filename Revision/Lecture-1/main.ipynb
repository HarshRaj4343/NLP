```python
import re
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/Users/harshraj/Desktop/NLP/questions.csv')
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df['question1'] = df['question1'].str.lower()
df['question2'] = df['question2'].str.lower()
```

```python
def remove_tags(text):
    cleaned = re.sub(r'<.*?>', '', text)
    return cleaned
```

```python
df['question1'] = df['question1'].apply(remove_tags)
df['question2'] = df['question2'].apply(remove_tags)
```

```python
df
```

```python
def remove_url(text):
    pattern = re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub(r'', text)
df['question1'] = df['question1'].apply(remove_url)
df['question2'] = df['question2'].apply(remove_url)
```

```python
import string,time
exclusion = string.punctuation
def remove_punc1(text):
    return text.translate(str.maketrans('', '', exclusion))


df['question1'] = df['question1'].apply(remove_punc1)
df['question2'] = df['question2'].apply(remove_punc1)
```

```python
from textblob import TextBlob

def spell_check(text):
    return str(TextBlob(text).correct()) 
df['question1'] = df['question1'].apply(spell_check)
df['question2'] = df['question2'].apply(spell_check)
```
