# HTML Stripper

```python
import re  #regex
def html_stripper(text):
    clean_text = re.sub(r'<.*?>', '', text)
    return clean_text

sample_text = "<p>Hello</p> this is a sample text with <b>HTML</b> tags."
clean_output = html_stripper(sample_text)
print(clean_output)
```

# Handling Emojis


```python
text_with_emojis = "This is a text with emojis 😃,😊"

# Encoding the text to normalize emojis into a machine-readable format
encoded_text = text_with_emojis.encode('utf-8')
print(encoded_text)
```

# Spell Checks and Corrections

```python
from textblob import TextBlob

# Sample text with multiple spelling errors
incorrect_text = "shrt and conditin during several generatins of modifed in same maner"

# Applying TextBlob's correct() function
corrected_text = TextBlob(incorrect_text).correct()
print(corrected_text)
```

# Tokenisations

```python
from nltk.tokenize import sent_tokenize, word_tokenize
import nltk

# Downloading the required NLTK data files (punkt)
nltk.download('punkt')

dummy_text = "This is a dummy text. It contains multiple sentences for tokenization."

# 1. Sentence Tokenization (Splits the paragraph into a list of sentences)
sentences = sent_tokenize(dummy_text)
print("Sentences:", sentences)

# 2. Word Tokenization (Splits the sentences into individual words)
for sentence in sentences:
    words = word_tokenize(sentence)
    print("Words:", words)
```
