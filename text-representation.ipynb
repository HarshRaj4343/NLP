# Bag Of Words

```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

# Example Dataset similar to the one used in the video
data = {
    'text': [
        "people watch campusx", 
        "watch campusx campusx", 
        "people write comment", 
        "campusx write comment"
    ]
}
df = pd.DataFrame(data)

# 1. Create the CountVectorizer Object
# You can pass hyper-parameters here, e.g. cv = CountVectorizer(binary=True, max_features=10)
cv = CountVectorizer()

# 2. Fit and Transform the text column
# Calling .toarray() is required because it returns a sparse matrix by default
bow_matrix = cv.fit_transform(df['text']).toarray()

# 3. View the Vocabulary
print("Vocabulary:", cv.vocabulary_)

# 4. View the Vectors
print("BoW Matrix:\n", bow_matrix)
```

# N-grams, Bi-grams, and Uni-grams

```python
from sklearn.feature_extraction.text import CountVectorizer

# 1. Unigrams and Bigrams together (ngram_range=(1, 2))
# This means vocabulary will include single words AND two-word phrases
cv_bigram = CountVectorizer(ngram_range=(1, 2))

bigram_matrix = cv_bigram.fit_transform(df['text']).toarray()

print("Vocabulary (Unigrams + Bigrams):", cv_bigram.vocabulary_)
print("Matrix:\n", bigram_matrix)

# 2. Only Bigrams (ngram_range=(2, 2))
cv_only_bigram = CountVectorizer(ngram_range=(2, 2))
only_bigram_matrix = cv_only_bigram.fit_transform(df['text']).toarray()
print("Vocabulary (Only Bigrams):", cv_only_bigram.vocabulary_)

# 3. Unigrams, Bigrams, and Trigrams together (ngram_range=(1, 3))
cv_trigram = CountVectorizer(ngram_range=(1, 3))
trigram_matrix = cv_trigram.fit_transform(df['text']).toarray()
print("Vocabulary (Up to Trigrams):", cv_trigram.vocabulary_)
```

# TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# 1. Create the TfidfVectorizer Object
tfidf = TfidfVectorizer()

# 2. Fit and Transform the text column
tfidf_matrix = tfidf.fit_transform(df['text']).toarray()

# 3. View the calculated TF-IDF matrix
print("TF-IDF Matrix:\n", tfidf_matrix)

# 4. View the IDF values calculated for the vocabulary
print("IDF Values:\n", tfidf.idf_)
print("Vocabulary mapping:", tfidf.vocabulary_)
```
